```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast

# mBART 모델 로드
model_name = "facebook/mbart-large-50-many-to-many-mmt"
tokenizer = MBart50TokenizerFast.from_pretrained(model_name)
model = MBartForConditionalGeneration.from_pretrained(model_name)

# 소스 및 타겟 언어 설정
tokenizer.src_lang = "en_XX"
target_lang = "ko_KR"

# 번역 함수
def translate_en_to_ko_mbart(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True)
    translated_tokens = model.generate(**inputs, forced_bos_token_id=tokenizer.lang_code_to_id[target_lang])
    return tokenizer.decode(translated_tokens[0], skip_special_tokens=True)

# 테스트
english_text = "What a wonderful day!"
print(translate_en_to_ko_mbart(english_text))
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# T5 모델 로드
model_name = "google/mt5-small"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# 번역 함수
def translate_en_to_ko_t5(text):
    input_text = f"translate English to Korean: {text}"
    inputs = tokenizer(input_text, return_tensors="pt", truncation=True)
    translated_tokens = model.generate(**inputs)
    return tokenizer.decode(translated_tokens[0], skip_special_tokens=True)

# 테스트
english_text = "Can you recommend a good book?"
print(translate_en_to_ko_t5(english_text))
```

```python
from transformers import pipeline

# 번역 파이프라인 생성
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-ko")

# 번역
english_text = "How are you?"
korean_translation = translator(english_text)
print(korean_translation[0]["translation_text"])
```
